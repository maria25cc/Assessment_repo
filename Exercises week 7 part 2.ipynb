# Practicing Pandas

## 2.8

In this we'll be investigating the carbon footprint of different foods. 

We'll be leveraging a dataset compiled by [Kasia Kulma](https://r-tastic.co.uk/post/from-messy-to-tidy/) and contributed to [R's Tidy Tuesday project](https://github.com/rfordatascience/tidytuesday).

Start by importing pandas with the alias `pd`.

````python
# Codes comes here
import pandas as pd
````

## 2.9 

The dataset we'll be working with has the following columns:

|column      |description |
|:-------------|:-----------|
|country       | Country Name |
|food_category | Food Category |
|consumption   | Consumption (kg/person/year) |
|co2_emmission | Co2 Emission (Kg CO2/person/year) |


Import the dataset as a dataframe named `df` from this url: <https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv>

````python
# Code comes here

url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv"

carbonfootprint_df = pd.read_csv(url)

carbonfootprint_df.head(10)
````

## 2.10 

How many rows and columns are there in the dataframe?

````python
#Code comes here
carbonfootprint_df.shape #the .shape function tells you the number of rows and coloums.
````

In this case, there are 1430 rows and 4 columns

## 2.11

What is the mean `co2_emission` of the whole dataset?

````python
# Code comes here
#the .mean value tells us the mean
mean_co2 = carbonfootprint_df["co2_emmission"].mean()
print("The mean CO2 emission of the dataset is:")
print(mean_co2)
````

## 2.12

What is the maximum `co2_emmission` in the dataset and which food type and country does it belong to?

````python
# Code comes here
#the .max function tells us the highest value
max_co2 = carbonfootprint_df["co2_emmission"].max()
print("The max CO2 emmissions of the dataset is:")
print(max_co2)
````

## 2.13

How many countries produce more than 1000 Kg CO2/person/year for at least one food type?

````python
#code comes here
#First we need to filter for countries with co2 emissions higher than 1000 in the dataset:
high_co2 = carbonfootprint_df[carbonfootprint_df["co2_emmission"] > 1000] 
print(high_co2)
````

````python
#Now we can the number of individual countries by using the .nunique function
num_countries = high_co2["country"].nunique()
print("The number of countries that produce over 1000 kg CO2/person for at least one food type is:")
print(num_countries)
````

## 2.14

Which country consumes the least amount of beef per person per year?


````python
#first I need to select only the rows for beef:
beef_df = carbonfootprint_df[carbonfootprint_df["food_category"] == "Beef"]

#now I need to find the lowest consumption:
lowest_beef = beef_df["consumption"].min()

#now I can find the country with the lowest beef consumption. 
country_min_beef = beef_df[beef_df["consumption"] == lowest_beef]["country"]

#now I can print the result
print("The country that consumes the lowest amount of beef is:")
print(country_min_beef.tolist())
````

## 2.15

What is the total emissions of all the meat products (Pork, Poultry, Fish, Lamb & Goat, Beef) in the dataset combined?

````python
# code comes here

#first I need to create a line to define what meat products are:
meat_products = ["Beef", "Pork", "Fish", "Poultry", "Lamb & Goat"]

#now I can filter the dataset for meat products. 
# I do this by creating a boolean mask for rows in the coloum food_category, filtering for meat_products.
meat_df = carbonfootprint_df[carbonfootprint_df["food_category"].isin(meat_products)]

# Now I can sum the CO2 emissions of the meat_df
total_meat_emissions = meat_df["co2_emmission"].sum()

#finally, I need to print the result.
print("Total meat emissions:")
print(total_meat_emissions)
````

# 2.16 

What is the total emissions of all other (non-meat) products in the dataset combined?

To do this exercise, we can subtract the number of total meat emissions, found in 2.15 from the total emissions for the dataset.


````python
#this line sums all the CO2 emmissions for all the food categories
Foodfootprint = carbonfootprint_df.groupby("food_category")["co2_emmission"].sum()
print(Foodfootprint)
````

````python
Foodfootprint.sum() #this lines sums all the rows from above in order to get a total. 
````

````python
#Now we can subtract the meat emissions from the total emissions to get the non-meat emissions
nonmeat_emissions = Foodfootprint.sum()-total_meat_emissions

print("The total non-meat CO2 emissions for the dataset:")
print(nonmeat_emissions)
````
